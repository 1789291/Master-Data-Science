## 1.9. Ejemplo gradient descent

````python
# Generamos nube de puntos

theta_0 = 2
theta_1 = 5
X = (np.random.randn(100) + 1) * 50
jitter = 50 * np.random.randn(100)
Y = theta_0 + theta_1 * X + jitter

plt.scatter(X,Y)
````

Descenso del gradiente

````python
# Inicializamos th0 y th1 (son las que intentaremos estimar)

th0 = 1
th1 = 1

# Calculamos el vector de estimaciones (h)

h = th0 + th1*X

# Calculamos el vector que representa las diferencias entre las estimaciones y los datos reales

difs = h - Y

# Definimos la función de coste y la evaluamos en th0 y th1

J = (1/len(X)) * (difs.sum())**2
````

### 2.2.6. Handling missing data

````python
week_data = pd.Series(['Lu', 'Ma', 'Mi', 'Ju', np.nan, np.nan, 'Do'])
week_data
````

Varias opciones: 
- Los sustituimos
- Los quitamos

````python
week_data.dropna()  # Me quita los registros con NaN.
````

````python
week_data[week_data != np.nan]

# Tras el corchete, se lee como un "tal que"

# No funciona. Las comparativas con not a number dan problemas.

# La manera es compararlo consigo mismo y que no sea
````

````python
np.nan == np.nan  # Así los podemos detectar
````

````python
week_data[week_data == week_data]
````

````python
# Otra manera:

week_data[week_data.isnull()]
````

````python
week_data[ ~ week_data.isnull()]  # Me quedo con "la negación" -> virgulilla
````

Para comprobar si nuestra serie tiene NaN, usaremos el comando isnull.

````python
# Otra manera:
week_data[week_data.notnull()]
````
