¿Qué es una red neuronal y por qué funciona tan bien?

Una red neuronal es una serie de operaciones matriciales hechas de manera recursiva.

Pongamos un ejemplo muy simple: tenemos una imagen de 3x3 píxeles. Una imagen está representada por un número en cada píxel. Eso es la entrada de la red. Tenemos en la salida el número que está en la imagen.

Tenemos dos capas dentro de la red neuronal. Entonces, con la entrada lo que hacemos son operaciones matriciales entre capa y capa. Tras cada salida de cada capa se va reduciendo el problema (la matriz) y al final tenemos el resultado final (por ejemplo, una matriz de 10x1 que nos representa qué número tenemos).

¿Qué hacemos en DL? Nosotros tenemos que definir la arquitectura del modelo y después aplicamos un algoritmo que nos encontrará el modelo para predecir eso. Ese algoritmo pondrá al principio unas matrices aleatorias, le pasamos la imagen, multiplica en las matrices, y obtiene la salida. Entonces, pongamos que he metido un 7 y como las matrices eran aleatorias sale un 1. El modelo sabe que tiene que ser un 7 (aprendizaje supervisado). Entonces se aplica un algoritmo (*backpropagation*), que teniendo en cuenta la diferencia en la salida actualiza los valores de las matrices de las capas y vuelve a aprender. Esto se hace en muchas pasadas.

Esto no es más que un problema de optimización: se van haciendo varias pasadas y se busca el error mínimo. El algoritmo de optimización suele ser el Descenso del Gradiente Estocástico. Hay diferentes metodos para hacer este gradiente (hay maneras de hacer el gradiente estocástico), y eso es otra cosa que tendremos que fijar.

Otro parámetro que tenemos que fijar es cómo inicializamos las matrices. De manera empírica se demuestra que dependiendo del modelo y de la aplicación, hay unas inicializaciones u otras que funcionan mejor.

¿Hay una regla fija? No. No hay un diagrama de flujo. En el diagrama hay algunas partes que están claras pero otras no.

¿Y por qué esto funciona? Lo que hace una red neuronal es aprender una nueva representación de los datos, lo que hace que el problema a solucionar sea mucho más sencillo.

In Deep Learning, we learn **layers** of increasingly better representations. The **deep** in Deep Learning refers to the number of layers that we learn. Before Deep Learning, the models that we are going to see today were much shallower. The most typical form of the layer representations is **neural networks**. As we will see, **neural networks** are just chain of matrix operations -- arrays of 2 dimensions--, or more precisely, tensor operations, we will operate in dimensions higher than 2.


* More about learning representations: https://www.bbvadata.com/not-deep-learning-learning-represent/

## Neural Networks

A neural network is just a stacked set of layers. Each layer is represented by a tensor, and it performs an affine transformation on each one of the layers.
* A matrix is a 2D tensor
* Affine transformation $\vec{y} = W\cdot\vec{x} + \vec{b}$, where $W$ is the weights tensor, $\vec{x}$ is the input data, $\vec{y}$ is the output data and $\vec{b}$ is called the bias
* $W$ and $\vec{b}$ are model parameters that need to be found by training

![](imgs/05_nn.png)

Note that this is a **linear operation**, therefore the result will be linear too. But the functions we want to learn are normally non-linear. How do we overcome this limiation? By using an **activation function**, that transforms the result from the linear operation. So, we actually do the following operation:

$\vec{y} = f\left(W\cdot\vec{x} + \vec{b}\right)$

There are several common activation functions, depending on the characteristics of the data and the kind of problem we want to solve. Here we will use the probably most often used activation function: **rectified linear unit**, or **relu** for friends:

![Relu plot](https://upload.wikimedia.org/wikipedia/commons/6/6c/Rectifier_and_softplus_functions.svg)

(image extracted from https://en.wikipedia.org/wiki/Rectifier_(neural_networks))

Relu will add the non-linearity that we need to learn arbitrary functions.

We can achieve the same with many other activation functions. See the documentation of Keras for details on the available activation functions:
 - https://keras.io/activations/


Each layer learns a new representation of the input data, and after several layers, the representation will make it straightforward to solve the task (e.g. assiging a digit to an image of a digit).

Until very recently, any attempt to train a deep neural network would not provide good results. Recent progresses (e.g. advances in hardware, new activation functions) make it possible to train very deep network, and therefore, solve problems that were unfeasible not long ago.

### Training process

The parameters of the network are initially randomly assigned. Then after each pass of the network, that is, after each *epoch*, the predictions of the network is compared against the true values, applying a *loss function*. This score is used to update the parameters of the network, with a process called **backpropagation**.
* Backpropagation is a complex process that involves calculating the gradient of the parameters of the network. For more details see: https://en.wikipedia.org/wiki/Backpropagation
* See also this excellent and visual explanation of how backpropagation works: https://google-developers.appspot.com/machine-learning/crash-course/backprop-scroll/

The training process is just an optimization process, that is normally done using a gradient descent-based algorithm. The values of gradients are computed using the backpropagation algorithm. Each epoch will make the loss score to be lower, and the predictions to be better.

![](imgs/06_optim.png)

In practice, we have to fine tune the number of epochs (and other parameters), until we obtain a result that is good enough for our purpose. In fact, if the number of epochs is too high, our model may **overfit** the training data. In that case, it will perform worsely when the model is exposed to data that it  has not seen previously.
* Fine tuning the number of epochs, and other parameters (e.g. the number and shape of the layers) is known as **hyperparameters optimization**, or **hyperparameters tuning**
* The parameters of the model are the $W$ and $b$ values of each layer, and those are determined by training.

## Hyperparameters tuning

In summary, when we are trying to obtain a deep learning model, the process that we will follow is:
* Decide on the architecture of the network (number of layers, type of each layer, size of each layer)
* Decide on the activation functions on each layer
* Decide on the parameters related to the optimization process (e.g. learning rate)
* Decide on the loss function that we will use
* Decide on the metrics that we will use to evaluate the performance of the model
* Decide on the training parameters (number of epochs, batch size)
* Decide on how to split the data between training, validation and test sets.

We obtain feedback for those decisions with the results of the training applied to the validation set. We train once, change some of the hyperparameters, train again, decide if that's better, and we keep going until we have found a model that is satisfactory enough for our purposes.

After that, we finally evaluate the model on the test set.

```python
%pylab inline
plt.style.use('seaborn-talk')
```

```python
# Esto es una función que vamos a utilizar para graficar

def plot_metric(history, metric):
    history_dict = history.history
    values = history_dict[metric]
    if 'val_' + metric in history_dict.keys():  
        val_values = history_dict['val_' + metric]

    epochs = range(1, len(values) + 1)

    if 'val_' + metric in history_dict.keys():  
        plt.plot(epochs, val_values, label='Validation')
    plt.semilogy(epochs, values, label='Training')

    if 'val_' + metric in history_dict.keys():  
        plt.title('Training and validation %s' % metric)
    else:
        plt.title('Training %s' % metric)
    plt.xlabel('Epochs')
    plt.ylabel(metric.capitalize())
    plt.legend()

    plt.show()  
```

### Preparar los datos

```python
x_train = np.linspace(0, 100, 1000)
y_train = 3*x_train + 2
```

```python
plot(x_train, y_train);
```

Este es el modelo que queremos predecir. Vamos a definir una red neuronal para intentar aprender de estos datos. 

El framework que utilizaremos será `keras`. `Keras` es una API de alto nivel (muy cercana) que va encima de TensorFlow (y otros frameworks) que lo hace un poquito más amigable.

`Keras` básicamente nos facilita la tarea con TensorFlow.

```python
# Importamos los modelos de keras
from keras import models

# Importamos las capas
from keras import layers
```

Definimos nuestro primer modelo:

Nota: hay muchísimos modelos. Nosotros sólo utilizaremos modelos secuenciales (una capa, y después otra, y en orden). El 95% de los modelos de DL son secuenciales.

```python
# Definimos el modelo
m = models.Sequential()
```

Una vez definido el modelo, podemos añadirle capas. También hay muchos tipos de capas. Nosotros utilizaremos capas densas (son matrices o tensores).

Tenemos que decidir el tamaño de la matriz. ¿Y qué tamaño es bueno? Pues ni idea. Habrá que ir ensayando.

Probamos con 64, lo que nos dirá que la matriz tendrá una dimensión de 64, y la otra en función de la que ncesite para que las operaciones matriciales puedan hacerse. Hemos puesto 64 por probar.

También tenemos que definir la forma de entrada de los datos. En este caso, con la regresión simple, la dimensión es una.

También hay que definir cómo será la salida. En este caso el problema es de regresión: tenemos un número, y queremos predecir otro número.

```python
# Primera capa
m.add(layers.Dense(64, input_shape=(1,)))  # dimensión 64, shape de entrada 1. Lo ponemos así porque la función come tuplas. Como es un escalar, es una tupla de un elemento

# Segunda capa: capa de salida
m.add(layers.Dense(1))
```

Lo que hemos hecho ha sido:

Matriz de un número ---> matriz de 64 por x ---> matriz de un número

Vamos a echar un vistazo al modelo:

```python
m.summary()
```

Tenemos **193 parámetros**. Son muchísimos. Una regresión lineal de primer grado tiene dos parámetros si lo hacemos por MCO. 

Hemos hecho un modelo súper simple y aún así tenemos un montón.

Con un modelo de DL el número de parámetros crece un montón, lo que nos llevará a tener overfitting. Lo veremos.

Vamos a ajustar el modelo. Hasta ahora sólo lo hemos definido. Tenemos que definir ahora:
* Una función de pérdidas
* Método de optimización
* Métricas de evaluación

```python
from keras import optimizers  # método de optimización
from keras import losses  # funciones de pérdida
from keras import metrics  # métricas
```

* Elegimos el optimizador `rmeprop`, que es de amplio uso. Es el SGD de toda la vida. También hay un método muy popular que es Adam. Aquí tenemos bastante manga ancha: elegir uno u otro no será crucial.

* La función de pérdida sí que es importante fijarla bien. La fijaremos en función del problema. En el caso de los problemas de regresión: MSE, RMSE, etc... En el caso de los modelos de clasificación, utilziaremos funciones de entropía cruzada. En este caso vamos a utilizar el error cuadrático medio.

* Métricas: le podemos meter una lista de las que queramos. Como cuando calcula la función de loss y su valor, y la función es el MSE, pues ahí ya tenemos una métrica. Podemos poner otras.

```python
# Definimos el modelo

m.compile(
          optimizer=optimizers.rmsprop(), 
          loss=losses.mean_squared_error,
          metrics=[])
```

Esto hay que hacerlo y estas decisiones las tenemos que tomar nosotros, junto con la arquitectura de la red.

¿Cómo es el proceso de encontrar la mejor red neuronal? Prueba y error.

Una vez definidos los hiperparámetros, ajustamos el modelo en sí. Hay que definir algún hiperparámetro más:

**Cantidad de pasadas**: o número de épocas. Una época es cuando todos los datos pasan por la red. Pero ojo:

- En Machine Learning lanzamos el GD y cuando ha llegado al mínimo o está lo suficientemente cerca para y nos da la respuesta. 
- En DL el tema de las épocas se hace por cuestiones prácticas. Con redes el ajuste puede llevarnos días, y a lo mejor pararnos un poco antes nos puede valer. Además, el entrenamiento es incremental: la red se va ajustando.

Otro hiperparámetro: **Batch size**: en vez de ir número a número en la red, agrupa lotes y pásaselos a la GPU. Cuanto más grande sea el batch size, más rápido será el ajuste, pero en función de la GPU. Pero perdemos granularidad si los lotes son grandes, porque la red neuronal sólo se ajusta a la salida. Si el tamaño es 2000, cada 2000 datos se actualizará la red. Digamos que si lo hacemos por batches hacemos "medias" en vez de ir uno a uno. Perdemos granularidad pero ganamos tiempo de cómputo. Y además, al ser medias tenemos menos varianza, lo que nos interesa.

La recomendación de los papers es que el batch size sea bastante grande como sea posible porque los resultados no van a ser tan malos y el tiempo va a ser poco.

También le vamos a decir que el 20% de los datos lo reserve para validar. Lo vemos.

```python
h = m.fit(x_train, y_train, epochs=10, batch_size=8192, validation_split=.2)  # h de histórico
```

Vamos a ver qué tal lo ha hecho el modelo:

```python
plot_metric(h, 'loss')
```

Tenemos **underfitting**. Vamos a modificar las épocas porque es obvio que podemos sacarle un poco más de rendimiento. Y por qué no tocamos otras cosas? Pues ya las tocaremos. De momento vamos a tocar las épocas que es evidente. Cuando ya no sea tan evidente, miramos

Ahora es incluso más evidente. Recordar que esto es escala logarítmica. Estamos viendo que el problema está yendo muy rápido hacia el mínimo, y me está saliendo barato (tardo muy poco en ejecutar), así que vamos a meterle 300 épocas.

¿Qué ha pasado? Aquí entran las funciones de activación.

**Una red neuronal sólo es capaz de predecir relaciones lineales** (un término $ x^2 $ no puede). De la forma $ y = Wx + b $. O sea, una red neuronal sólo puede aprender transformaciones afines. Es decir, en nuestro caso de 2d líneas. Además se demuestra que muchas capas (muchas matrices) combinadas al final siguen dando una transformación afín.

Si queremos predecir otras funciones, necesitamos las funciones de activación, de manera que sea $ y = f(Wx +b) $.Mediante estas funciones le metemos no linealidad al modelo.

Vamos a meter la función de activacitón `relu`, que nos da 0 si el valor es negativo y 1 si es positivo.

```python
m = models.Sequential()

m.add(layers.Dense(128, input_shape=(1,), activation='relu')) 
m.add(layers.Dense(64, activation='relu'))
m.add(layers.Dense(1))  # Aquí no le metemos función de activación porque en la salida no queremos restringir a 0 o 1

m.compile(
          optimizer=optimizers.rmsprop(), 
          loss=losses.mean_squared_error,
          metrics=[])

h = m.fit(x_train, y_train, epochs=1000, batch_size=8192, validation_split=.2, verbose=0)

plot_metric(h, 'loss')
```

```python
y_pred = m.predict(x_train)
plt.plot(x_train, y_pred, 'ro')
plt.plot(x_train, y_train)
```

¿Qué hacemos? Podemos, por ejemplo, meter más neuronas.

```python
m = models.Sequential()

m.add(layers.Dense(256, input_shape=(1,), activation='relu')) 
m.add(layers.Dense(128, activation='sigmoid')) 
m.add(layers.Dense(64, activation='relu'))
m.add(layers.Dense(1))  # Aquí no le metemos función de activación porque en la salida no queremos restringir a 0 o 1

m.compile(
          optimizer=optimizers.rmsprop(), 
          loss=losses.mean_squared_error,
          metrics=[])

h = m.fit(x_train, y_train, epochs=1000, batch_size=8192, validation_split=.2, verbose=0)

plot_metric(h, 'loss')

y_pred = m.predict(x_train)
plt.plot(x_train, y_pred, 'ro')
plt.plot(x_train, y_train)
```

Vemos que tampoco. Al final, probando, conseguiríamos un mejor o peor fit.

Otra cosa que también funciona es normalizar los datos. Esto ayuda porque, por ejemplo, si los números son muy grandes, hay más problemas en los cálculos numéricos y en sus aproximaciones.
